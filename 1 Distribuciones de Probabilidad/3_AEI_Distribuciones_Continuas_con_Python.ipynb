###**DISTRIBUCIÓN UNIFORME**



Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.  

Entonces, definimos la variable aleatoria X=“aquella que solo puede tomar valores entre a y b” que sigue una distribución Uniforme
\begin{align}
X \sim U(a,b)
\end{align}

con a (límite inferior del intervalo) y b( límite superior del intervalo) los parámetros de la distribución

Función de densidad de probabilidad:

\begin{align}
f(x) = {1\over{b-a}} \quad\mathit{si}\quad a \leq x \leq b
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)= {x-a\over{b-a}} \quad\mathit{si}\quad a \leq x \leq b
\end{align}

Valor esperado:
\begin{align}
E(X)={a+b\over{2}}
\end{align}

Varianza:
\begin{align}
V(X)={(b-a)^2\over{12}}
\end{align}

Por ejemplo: sea ${X \sim U(3,5)}$. Calcule la probabilidad F(4)

```python
round(ss.uniform.cdf(4,loc=3, scale=5-3),4)
```

Se verifica este resultado del siguiente modo:

${F(4)=} {{\int_{3}^{4}}{1\over (5-3)} dx}$

${F(4)=} {{\int_{3}^{4}}{1\over 2} dx}$

Resolviendo

${F(4) =} {{1\over2}x \Big|_{3}^4}$

${F(4) = 0.5}$

Comportamiento de la distribución Uniforme:

```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import numpy as np

#Parámetros de la distribución Uniforme
a,b = 0,1
a_1,b_1 = 1,3
a_2,b_2 = 2,5

#Valores de X
x = np.arange(0,1,0.1)
x_1 = np.arange(1,3,0.1)
x_2 = np.arange(2,5,0.1)

#Densidad de probabilidad
fmp = ss.uniform.pdf(x, a, b-a)
fmp_1 = ss.uniform.pdf(x_1, a_1, b_1-a_1)
fmp_2 = ss.uniform.pdf(x_2, a_2, b_2-a_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(25,15))

#Gráfico de densidad de probabilidad
ax1.plot(x, fmp, color='blue', label='Parámetros: a=0 b=1')
ax1.set_xticks(x)
ax1.set_ylim(0,1.01)
ax1.set_title('Densidad de Probabilidad')
ax1.set_ylabel('Probabilidad')
ax1.set_xlabel('valores de X')
ax1.legend(fontsize=10)

# Distribución de Probabilidad
cdf = ss.uniform.cdf(x, a,b-a)
cdf_1 = ss.uniform.cdf(x_1, a_1, b_1-a_1)
cdf_2 = ss.uniform.cdf(x_2, a_2, b_2-a_2)

#Gráfico de distribución de probabilidad
ax2.plot(x, cdf,color='blue', label='Parámetros: a=0 b=1')
ax2.set_xticks(x)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores de X')
ax2.legend(fontsize=10)

ax3.plot(x_1, fmp_1, color='green', label='Parámetros: a=1 b=3')
ax3.set_xticks(x_1)
ax3.set_title('Densidad de Probabilidad')
ax3.set_ylabel('Probabilidad')
ax3.set_xlabel('valores de X')
ax3.legend(fontsize=10)

ax4.plot(x_1,cdf_1, color='green', label='Parámetros: a=1 b=3')
ax4.set_xticks(x_1)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores de X')
ax4.legend(fontsize=10)

ax5.plot(x_2, fmp_2, color='red', label='Parámetros: a=2 b=5')
ax5.set_xticks(x_2)
ax5.set_title('Densidad de Probabilidad')
ax5.set_ylabel('Probabilidad')
ax5.set_xlabel('valores de X')
ax5.legend(fontsize=10)

ax6.plot(x_2, cdf_2, color='red', label='Parámetros: a=2 b=5')
ax6.set_xticks(x_2)
ax6.set_ylim(0,1)
ax6.set_title('Distribución de Probabilidad')
ax6.set_ylabel('F(x)')
ax6.set_xlabel('valores de X')
ax6.legend(fontsize=10)

plt.show()
```

###**DISTRIBUCIÓN NORMAL**


Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.  

Entonces, definimos la variable aleatoria X=“aquella que solo puede tomar valores entre $-\infty$ y $+\infty$ que sigue una distribución Normal
\begin{align}
X \sim N(\mu,\sigma)
\end{align}

con $\mu$ (media) y $\sigma$ (desvío estandar) los parámetros de la distribución

Cuando $\mu=0$ y $\sigma=1$ la distribución recibe el nombre de Normal Estandar

Función de densidad de probabilidad:

\begin{align}
f(x) = {1\over \sigma*\sqrt{2 \pi}} e^{-{1\over2}{({(x-\mu)\over\sigma})}^2}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)= {\int_{-\infty}^{X}}{1\over \sigma*\sqrt{2 \pi}} e^{-{1\over2}{({(x-\mu)\over\sigma})}^2} dx
\end{align}

Valor esperado:
\begin{align}
E(X)= \mu
\end{align}

Varianza:
\begin{align}
V(X)= \sigma^2
\end{align}

Por ejemplo: sea ${X \sim N(2,1)}$. Calcule la probabilidad F(4)

```python
round(ss.norm.cdf(4, 2, 1),4)
```

Se verifica este resultado del siguiente modo:

${F(4)= {\int_{-\infty}^{4}}{1\over \sigma*\sqrt{2 \pi}} e^{-{1\over2}{({(x-\mu)\over\sigma})}^2} dx}$

${F(4)= {\int_{-\infty}^{4}}{1\over \sigma*\sqrt{2 \pi}} e^-{(x-\mu)^2 \over2\sigma^2} dx}$

Dado que ${\mu=2}$ y ${\sigma=1}$

${F(4)= {\int_{-\infty}^{4}}{1\over 1*\sqrt{2 \pi}} e^{-{(x-2)^2\over2}} dx}$

Resolviendo

${F(4) \approx 0.9772}$

Comportamiento de la distribución Normal:

```python
#Parámetros de la distribución Normal
mu, sigma = 0, 1
mu_1, sigma_1 = -2, 2
mu_2, sigma_2 = 2, 4

#Valores de X
x = np.arange(-15,15,0.5)
x_1 = np.arange(-15,15,1)
x_2 = np.arange(-15,15,1)

#Densidad de probabilidad
fmp = ss.norm.pdf(x, mu, sigma)
fmp_1 = ss.norm.pdf(x_1, mu_1, sigma_1)
fmp_2 = ss.norm.pdf(x_2, mu_2, sigma_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(35,25))

#Gráfico de Densidad de probabilidad
ax1.plot(x, fmp, color='blue', label='Parámetros: mu=0  sigma=1')
ax1.set_xticks(x)
ax1.set_ylim(0,1)
ax1.set_title('Densidad de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('valores')
ax1.legend(fontsize=10)


# Distribución de probabilidad
cdf = ss.norm.cdf(x, mu, sigma)
cdf_1 = ss.norm.cdf(x_1, mu_1, sigma_1)
cdf_2 = ss.norm.cdf(x_2, mu_2, sigma_2)

#Gráfico de distribución de probabilidad
ax2.plot(x, cdf, color='blue', label='Parámetros: mu=0  sigma=1')
ax2.set_xticks(x)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores')
ax2.legend(fontsize=10)


ax3.plot(x_1, fmp_1, color='green', label='Parámetros: mu=-2  sigma=2')
ax3.set_xticks(x_1)
ax3.set_ylim(0,1)
ax3.set_title('Densidad de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('valores')
ax3.legend(fontsize=10)

ax4.plot(x_1, cdf_1, color='green', label='Parámetros: mu=-2  sigma=2')
ax4.set_xticks(x_1)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores')
ax4.legend(fontsize=10)

ax5.plot(x_2, fmp_2, color='red', label='Parámetros: mu=2  sigma=4')
ax5.set_xticks(x_2)
ax5.set_ylim(0,1)
ax5.set_title('Densidad de Probabilidad')
ax5.set_ylabel('probabilidad')
ax5.set_xlabel('valores')
ax5.legend(fontsize=10)

ax6.plot(x_2, cdf_2, color='red', label='Parámetros: mu=2  sigma=4')
ax6.set_xticks(x_1)
ax6.set_ylim(0,1)
ax6.set_title('Distribución de Probabilidad')
ax6.set_ylabel('F(x)')
ax6.set_xlabel('valores')
ax6.legend(fontsize=10)

plt.show()
```

###**DISTRIBUCIÓN GAMMA**




Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.

Entonces, definimos la variable aleatoria X=“aquella que solo puede tomar valores entre $0$ y ${+\infty}$ que sigue una distribución Gamma
\begin{align}
X \sim \Gamma(\alpha,\beta)
\end{align}

con ${\alpha}$ (escala) y ${\beta}$ (forma) los parámetros de la distribución

Función de densidad de probabilidad:

\begin{align}
{f(x) =} {{(x\alpha)^{\beta}} \over {x\Gamma(\beta)}} { e^{-{x\alpha}}} {\thinspace} donde {\thinspace} x>0, {\alpha >0, \beta >0}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)={\int_{0}^{X}} {{(x\alpha)^{\beta}} \over {x\Gamma(\beta)}} { e^{-{x\alpha}}} dx
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {\beta \over \alpha}
\end{align}

Varianza:
\begin{align}
{V(X)=} {\beta \over \alpha^2}
\end{align}

Por ejemplo: sea $X \sim Ga(1, 2)$. Calcule la probabilidad F(4)

NOTA: la función de densidad gamma programada en esta librería es:

${f(x) = {{{x^{(a-1)}} {e^{-x}}} \over \Gamma(a)}}$

donde $a= \beta$ y posee parámetros loc=0 y scale=1 con $scale = 1/\alpha$

Si ${X \sim Ga(1, 2)}$ su equivalente entonces es $a= 2$ y por default loc=0 y scale=1 con $scale = 1/1=1$

Si ${X \sim Ga(2, 3)}$ su equivalente entonces es $a= 3$ y loc=0 y $scale = 1/2$

y así sucesivamente


```python
round(ss.gamma.cdf(4,2),4)
```

Se verifica este resultado del siguiente modo:



${F(4)={\int_{0}^{X}} {{(x\alpha)^{\beta}} \over {x\Gamma(\beta)}} { e^{-{x\alpha}}} dx}$

Dado que $\alpha=1$ y $\beta=2$

${F(4)={\int_{0}^{X}} {{(x*1)^{2}} \over {x\Gamma(2)}} { e^{-{x/1}}} dx}$

Resolviendo

${F(4)={(-x-1)}e^{-x} + 1 \Big|_{0}^4}$

${F(4)={((-4-1)e^{-4} + 1)}  - {((-0-1)e^{-0} + 1)}}$

${F(4)={(-5e^{-4} + 1)}  - {(-1e^{-0} + 1)}}$

${F(4)=0.908421805  - 0}$

${F(4) \approx  0.9084}$

Comportamiento de la distribución Gamma:

```python
#Parámetros de la distribución Gamma
a,alpha = 2,1
a_1,alpha_1 = 6,1
a_2,alpha_2 = 1,1

#valores de X
x = np.arange(0,10,0.2)
x_1 = np.arange(0,15,0.5)
x_2 = np.arange(0,10,0.2)

#Densidad de Probabilidad
fmp = ss.gamma.pdf(x, a, loc=0, scale=1/alpha)
fmp_1 = ss.gamma.pdf(x_1, a_1, loc=0, scale=1/alpha_1)
fmp_2 = ss.gamma.pdf(x_2, a_2, loc=0, scale=1/alpha_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(35,25))

#Gráfico de densidad de probabilidad
ax1.plot(x, fmp,  color='blue', label='Parámetros: alpha=1, beta=2')
ax1.set_xticks(x)
ax1.set_ylim(0,1)
ax1.set_title('Densidad de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('valores de X')
ax1.legend(fontsize=10)

# Distribución de probabilidad
cdf = ss.gamma.cdf(x, a, loc=0, scale=1/alpha)
cdf_1 = ss.gamma.cdf(x_1, a_1, loc=0, scale=1/alpha_1)
cdf_2 = ss.gamma.cdf(x_2, a_2, loc=0, scale=1/alpha_2)

#Gráfico de distribución de probabilidad
ax2.plot(x, cdf, color='blue', label='Parámetros: alpha=1, beta=2')
ax2.set_xticks(x)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores de X')
ax2.legend(fontsize=10)

ax3.plot(x_1, fmp_1, color='green', label='Parámetros: alpha=1, beta=6')
ax3.set_xticks(x_1)
ax3.set_ylim(0,1)
ax3.set_title('Densidad de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('valores de X')
ax3.legend(fontsize=10)

ax4.plot(x_1, cdf_1, color='red', label='Parámetros: alpha=1, beta=6')
ax4.set_xticks(x_1)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores de X')
ax4.legend(fontsize=10)

ax5.plot(x_2, fmp_2, color='red', label='Parámetros: alpha=1, beta=1')
ax5.set_xticks(x_2)
ax5.set_ylim(0,1)
ax5.set_title('Densidad de Probabilidad')
ax5.set_ylabel('probabilidad')
ax5.set_xlabel('valores de X')
ax5.legend(fontsize=10)

ax6.plot(x_2, cdf_2, color='green', label='Parámetros: alpha=1, beta=1')
ax6.set_xticks(x_2)
ax6.set_ylim(0,1)
ax6.set_title('Distribución de Probabilidad')
ax6.set_ylabel('F(x)')
ax6.set_xlabel('valores de X')
ax6.legend(fontsize=10)

plt.show()
```

###**DISTRIBUCIÓN EXPONENCIAL**



Se caracteriza por determinar la probabilidad de ocurrencia de la variable aletoria en un intervalo de valor pero a una tasa de ocurrencia constante en el tiempo o el espacio .

Entonces, definimos la variable aleatoria X=“aquella que solo puede tomar valores entre $0$ y ${+\infty}$ a una tasa constante en el tiempo o el espacio" que sigue una distribución Exponencial
\begin{align}
X \sim Exp(\alpha)
\end{align}

con ${\alpha}$ (tasa constante en el tiempo o el espacio) el parámetro de la distribución

Función de densidad de probabilidad:

\begin{align}
{f(x) =} {\alpha e^{-\alpha x}} {\thinspace}{donde} {\thinspace}{x>0}, {\alpha} = {1\over λ}
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)={ 1-e^{-{\alpha x}}}
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {1 \over \alpha}
\end{align}

Varianza:
\begin{align}
{V(X)=} {1 \over \alpha^2}
\end{align}

Por ejemplo: sea $X \sim Exp(0.5)$. Calcule la probabilidad P(X<=2)

NOTA: la función de densidad Exponencial programada en esta librería es:

${f(x) = e^{-x}}$

que posee parámetros loc=0 y scale=1 con ${scale = 1/\alpha}$

Si ${X \sim Exp(0.5)}$ su equivalente entonces es loc=0 y scale=2 con ${scale = 1/0.5=2}$

```python
round(ss.expon.cdf(2, loc=0, scale=2),4)
```

Se verifica este resultado del siguiente modo:


${F(2)={\int_{0}^{2}} {{\alpha}e^{-{\alpha}x}} dx}$

Dado que $\alpha=0.5$

${F(2)={\int_{0}^{2}} {{0.5}e^{-{0.5}x}} dx}$

Resolviendo

${F(2)= -e^{-x \over2} \Big|_{0}^2}$

${F(2)= -e^{-2 \over2} - (-e^{-0 \over2})}$

${F(2)= -0.367879441 + 1}$

${F(2) \approx  0.6321}$

Comportamiento de la distribución Exponencial:

```python
#Parámetros de la distribución Exponencial
alpha = 0.2
alpha_1 = 0.5
alpha_2 = 0.8

#valores de X
x = np.arange(0.1, 15, 0.5)
x_1 = np.arange(0.1, 15, 0.5)
x_2 = np.arange(0.1, 15, 0.5)

#Densidad de Probabilidad
fmp = ss.expon.pdf(x, loc=0, scale=1/alpha)
fmp_1 = ss.expon.pdf(x_1, loc=0, scale=1/alpha_1)
fmp_2 = ss.expon.pdf(x_2, loc=0, scale=1/alpha_2)

#parámetros gráficos
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(35,25))

#Gráfico de densidad de probabilidad
ax1.plot(x, fmp,  color='blue', label='Parámetros: alpha=0.2')
ax1.set_xticks(x)
#ax1.set_xlim(0,20)
ax1.set_ylim(0,1)
ax1.set_title('Densidad de Probabilidad')
ax1.set_ylabel('probabilidad')
ax1.set_xlabel('valores de x')
ax1.legend(fontsize=10)

# Distribución de probabilidad
cdf = ss.expon.cdf(x, loc=0, scale=1/alpha)
cdf_1 = ss.expon.cdf(x_1, loc=0, scale=1/alpha_1)
cdf_2 = ss.expon.cdf(x_2, loc=0, scale=1/alpha_2)

#Gráfico de distribución de probabilidad
ax2.plot(x, cdf, color='blue', label='Parámetros: alpha=0.2')
ax2.set_xticks(x)
#ax2.set_xlim(0,40)
ax2.set_ylim(0,1)
ax2.set_title('Distribución de Probabilidad')
ax2.set_ylabel('F(x)')
ax2.set_xlabel('valores de x')
ax2.legend(fontsize=10)


ax3.plot(x_1, fmp_1, color='red', label='Parámetro: alpha=0.5')
ax3.set_xticks(x_1)
#ax3.set_xlim(0,20)
ax3.set_ylim(0,1)
ax3.set_title('Densidad de Probabilidad')
ax3.set_ylabel('probabilidad')
ax3.set_xlabel('valores de x')
ax3.legend(fontsize=10)

ax4.plot(x_1, cdf_1, color='green', label='Parámetro: alpha=0.5')
ax4.set_xticks(x_1)
#ax4.set_xlim(0,40)
ax4.set_ylim(0,1)
ax4.set_title('Distribución de Probabilidad')
ax4.set_ylabel('F(x)')
ax4.set_xlabel('valores de x')
ax4.legend(fontsize=10)

ax5.plot(x_2, fmp_2, color='green', label='Parámetro: alpha=0.8')
ax5.set_xticks(x_2)
#ax5.set_xlim(0,20)
ax5.set_ylim(0,1)
ax5.set_title('Densidad de Probabilidad')
ax5.set_ylabel('probabilidad')
ax5.set_xlabel('valores de x')
ax5.legend(fontsize=10)

ax6.plot(x_2, cdf_2, color='red', label='Parámetro: alpha=0.8')
ax6.set_xticks(x_2)
#ax6.set_xlim(0,40)
ax6.set_ylim(0,1)
ax6.set_title('Distribución de Probabilidad')
ax6.set_ylabel('F(x)')
ax6.set_xlabel('valores de x')
ax6.legend(fontsize=10)

plt.show()
```

###**DISTRIBUCIÓN BETA**



Se caracteriza por determinar la probabilidad de ocurrencia entre los extremos 0 y 1 de la variable aletoria, es decir, en un intervalo de valor.

Entonces, definimos la variable aleatoria X=“aquella que solo puede tomar valores entre $0$ y $1$ que sigue una distribución Beta
\begin{align}
X \sim Beta(a,b)
\end{align}

con ${a}$ (escala) y ${b}$ (forma) los parámetros de la distribución



Función de densidad de probabilidad:

\begin{align}
{f(x) =} {{{x^{a-1}} {(1-x)^{b-1}}}\over {B(a;b)}} {\thinspace} donde {\thinspace} 0<x<1, {a >0, b >0}
\end{align}

Con $B(a;b)={\int_{0}^{1} {{{x^{a-1}} {(1-x)^{b-1}}}}}$

Función de distribución de probabilidad:

\begin{align}
F(X)={1\over {B(a;b)}}{\int_{0}^{X}} {{(x)^{a-1}} {(1-x)^{b-1}}} dx
\end{align}

Valor esperado:
\begin{align}
{E(X)= \mu=} {a \over (a+b)}
\end{align}

Varianza:
\begin{align}
{V(X)=} {\mu (1-\mu) \over (a+b+1)}
\end{align}

Por ejemplo: sea $X \sim Beta(1, 2)$. Calcule la probabilidad F(0.5)

```python
round(ss.beta.cdf(0.5,1,2),4)
```

Se verifica este resultado del siguiente modo:

${F(0.5)={1\over {B(a;b)}}{\int_{0}^{X}} {{(x)^{a-1}} {(1-x)^{b-1}}} dx}$

Dado que $a=1$ y $b=2$

${F(0.5)={1\over {B(1;2)}}{\int_{0}^{X}} {{(x)^{1-1}} {(1-x)^{2-1}}} dx}$

Resolviendo

${F(0.5)={1\over {B(1;2)}}{\int_{0}^{0.5}} {{(1-x)^{1}}} dx}$

${F(0.5)={1\over {B(1;2)}}[{\int_{0}^{0.5}} {1 dx} - {\int_{0}^{0.5}} {x dx}] }$

$B(1;2)={\int_{0}^{1} {{{x^{a-1}} {(1-x)^{b-1}}}}}={1\over 2}$

${F(0.5)=2*[{0.5}  - {0.125}}]$

${F(0.5)=2*[{0.375}]}$

${F(0.5)=0.75}$

###**DISTRIBUCIÓN T-STUDENT**




Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.

Entonces, definimos la variable aleatoria X=“aquella que puede tomar valores entre ${-\infty}$ y ${+\infty}$ " que sigue una distribución T-Student
\begin{align}
X \sim t_n
\end{align}

con $n$ (grados de libertad) el párametro de la distribución

Función de densidad de probabilidad:

\begin{align}
{f(x) =} {{{\Gamma({n+1\over2})} \over {\Gamma({n\over2})*\sqrt{n \pi}}}} *
{({1+{ x^2\over n})}^{-{{n+1}\over2}}} {\thinspace} donde {\thinspace} n>0
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)={\int_{-\infty}^{X}} {{{\Gamma({n+1\over2})} \over {\Gamma({n\over2})*\sqrt{n \pi}}}} *
{({1+{ x^2\over n})}^{-{{n+1}\over2}}} dx
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {0}
\end{align}

Varianza:
\begin{align}
{V(X)=} {n \over n-2}
\end{align}

Por ejemplo: sea $X \sim t_8$. Calcule la probabilidad F(2)

```python
round(ss.t.cdf(2,8),4)
```

Se verifica este resultado del siguiente modo:

${F(2)={\int_{-\infty}^{X}} {{{\Gamma({n+1\over2})} \over {\Gamma({n\over2})*\sqrt{n \pi}}}} *
{({1+{ x^2\over n})}^{-{{n+1}\over2}}} dx}$

Dado que $n=8$

${F(2)={\int_{-\infty}^{2}} {{{\Gamma({8+1\over2})} \over {\Gamma({8\over2})*\sqrt{8 \pi}}}} *
{({1+{ x^2\over 8})}^{-{{8+1}\over2}}} dx}$

${F(2)={{{\Gamma({8+1\over2})} \over {\Gamma({8\over2})*\sqrt{8 \pi}}}} {\int_{-\infty}^{2}}  
{({1+{ x^2\over 8})}^{-{4.5}}} dx}$

${F(2)={0.386699021} {\int_{-\infty}^{2}}  
{({1+{ x^2\over 8})}^{-{4.5}}} dx}$

Resolviendo

${F(2)={0.386699021} * 2.481883400261275}$

${F(2) \approx  0.9597}$

###**DISTRIBUCIÓN CHI-CUADRADO**




Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.

Entonces, definimos la variable aleatoria X=“aquella que puede tomar valores entre $0$ y ${+\infty}$ " que sigue una distribución Chi-Cuadrado
\begin{align}
X \sim {\chi^2}_n
\end{align}

con $n$ (grados de libertad) el párametro de la distribución

Función de densidad de probabilidad:

\begin{align}
{f(x) =} {{x^{({n\over2}-1)}} e^{-x\over2} \over {2^{n\over2}{\Gamma({n\over2})}}} {\thinspace} donde {\thinspace} n>0
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)= {\int_{0}^{X}}{{x^{({n\over2}-1)}} e^{-x\over2} \over {2^{n\over2}{\Gamma({n\over2})}}} dx
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {n}
\end{align}

Varianza:
\begin{align}
{V(X)=} {2n}
\end{align}

Por ejemplo: sea $X \sim {\chi^2}_8$. Calcule la probabilidad F(2)

```python
round(ss.chi2.cdf(2,8),4)
```

Se verifica este resultado del siguiente modo:

${F(2)={\int_{0}^{X}}{{x^{({n\over2}-1)}} e^{-x\over2} \over {2^{n\over2}{\Gamma({n\over2})}}} dx}$

Dado que $n=8$

${F(2)={\int_{0}^{2}}{{x^{({8\over2}-1)}} e^{-x\over2} \over {2^{8\over2}{\Gamma({8\over2})}}} dx}$

${F(2)={\int_{0}^{2}}{{x^{(3)}} e^{-x\over2} \over {2^{4}{\Gamma({4})}}} dx}$

${F(2)={1\over96} {\int_{0}^{2} {x^3} {e^{-x\over2}}} dx}$

Resolviendo

${F(2)={1\over96}*1.822863060110766}$

${F(2) \approx  0.019}$

###**DISTRIBUCIÓN F-SNEDECOR**




Se caracteriza por determinar la probabilidad de ocurrencia entre dos extremos de la variable aletoria, es decir, en un intervalo de valor.

Entonces, definimos la variable aleatoria X=“aquella que puede tomar valores entre $0$ y ${+\infty}$ " que sigue una distribución F-Snedecor
\begin{align}
X \sim {F}_{n_1, n_2}
\end{align}

con $n_1$ (grados de libertad del numerador) y $n_2$ (grados de libertad del denominador) los párametros de la distribución

Función de densidad de probabilidad:

\begin{align}
{f(x) =} {{{\Gamma({n_1 +n_2\over2})} \over {\Gamma({n_1\over2})}{\Gamma({n_2\over2})}} {({n_1\over n_2})}^{n_1\over 2} {{x^{{n_1\over2}-1}} \over
{({1+ {n_1x\over n_2}})}^{{n_1+n_2}\over2} } }
\end{align}

Función de distribución de probabilidad:

\begin{align}
F(X)= {\int_{0}^{X}}{{{\Gamma({n_1 +n_2\over2})} \over {\Gamma({n_1\over2})}{\Gamma({n_2\over2})}} {({n_1\over n_2})}^{n_1\over 2} {{x^{{n_1\over2}-1}} \over
{({1+ {n_1x\over n_2}})}^{{n_1+n_2}\over2} } } dx
\end{align}

Valor esperado:
\begin{align}
{E(X)=} {n_2\over{n_2-2}} {\thinspace} con {\thinspace} {n_2}>2
\end{align}

Varianza:
\begin{align}
{V(X)=} {2*n_2^2 *{(n_1+n_2-2)}\over{n_1*{(n_2-2)}^2 *{(n_2-4)}}} {\thinspace} con {\thinspace} {n_2}>4
\end{align}

Por ejemplo: sea $X \sim F_{7,10}$. Calcule la probabilidad F(2)

```python
round(ss.f.cdf(2,7,10),4)
```

${F(2)={\int_{0}^{X}}{{{\Gamma({n_1 +n_2\over2})} \over {\Gamma({n_1\over2})}{\Gamma({n_2\over2})}} {({n_1\over n_2})}^{n_1\over 2} {{x^{{n_1\over2}-1}} \over
{({1+ {n_1x\over n_2}})}^{{n_1+n_2}\over2} } } dx}$

Dado que $n_1=7$ y $n_2=10$

${F(2)={\int_{0}^{2}}{{{\Gamma({7 +10\over2})} \over {\Gamma({7\over2})}{\Gamma({10\over2})}} {({7\over 10})}^{7\over 2} {{x^{{7\over2}-1}} \over
{({1+ {7x\over 10}})}^{{7+10}\over2} } } dx}$

${F(2)={{\Gamma({7 +10\over2})} \over {\Gamma({7\over2})}{\Gamma({10\over2})}} {({7\over 10})}^{7\over 2}{\int_{0}^{2}} {{x^{{7\over2}-1}} \over
{({1+ {7x\over 10}})}^{{7+10}\over2}} dx}$

Resolviendo

$F(2)={50.4951615510274} * {0.01674411317357969}$

${F(2) \approx  0.8455}$
