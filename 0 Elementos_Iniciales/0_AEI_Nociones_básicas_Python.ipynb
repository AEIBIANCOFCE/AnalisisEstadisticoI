##**Condicional IF**

La sentencia IF permite instruir acciones alternativas a realizar sobre un conjunto de resultados posibles.

La sintaxis general es

    if condición1:
        realizá la acción1
    elif condición2:
        realizá la acción3
    else:
        realizá la acción3

Comenzemos con una instrucción simple

```python
#definimos una variable sobre la cual operar
x = True

if x==True:
  print('x es verdadero!')
```

```python
x = False

if x==True:
    print('x es verdadero!')
```

Si no se cumple la condición no devuelve nada. Entonces, podemos completar la alternativa del siguiente modo

```python
x = False

if x==True:
    print('x es verdadero!')
else:
    print('x no es verdadero!')
```

También podría ser util tener una función que cuente las veces que aparece una palabra en un campo de mis datos

```python
def buscar_palabra(lista, palabra):
    #cuenta las veces que aparece una palabra (o podría pensarse en buscar un producto, etc)
    for n,i in enumerate(lista):
      if i ==palabra :
       result=lista.count(palabra)
    return result
```

Convertimos en función el ejemplo 3

```python
def num_par(lista_num):
  '''
  Función que verifica si un número es par
  Parámetros:
    lista_num: lista de números
  Devuelve: si cada número es par o impar
  '''
  if type(lista_num)==str:
    numeros=lista_num.split(',')
    numeros=list(map(int, numeros))
  elif type(lista_num)==list:
    numeros=lista_num
  else:
    print('El tipo de dato ingresado no es válido')
  resultado=[]
  for i in numeros:
    if i >= 0 and i%2==0:
      resultado.append('El número '+str(i)+' es par')
    elif  i>= 0 and i%2!=0:
      resultado.append('El número '+str(i)+' es impar')
    else:
      resultado.append('El número '+str(i)+' no corresponde')
  return resultado
```

Convertimos a función el ejemplo 4

```python
def capitalizacion(montoinicial, interes, años):
  '''
  Función que calcula la capitalización de un monto invertido
  Parámetros:
    montoinicial: monto a invertir
    interes: tasa de interés anual
    años: números de años a invertir
  Devuelve: la capitalización de un monto invertido año por año y si se alcanzó el límite de capitalización
  '''
  respuesta=[]
  montoacum=montoinicial
  while montoinicial >0 and interes>0 and años>0 and montoacum < montoinicial*2:
    for i in range(años):
      montoacum = montoacum +(montoacum * interes / 100)
      if montoacum < montoinicial*2:
        respuesta.append("Capital tras " + str(i+1) + " años: " + str(round(montoacum, 2)))
      else:
        respuesta.append('Se alcanzó el límite de capitalización')
        break
  else:
    respuesta.append('Los valores ingresados son incorrectos')
  return respuesta
```
