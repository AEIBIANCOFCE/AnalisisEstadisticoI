Paso 4:

```python
#importar librerías
import requests
import pandas as pd
```

```python
#seteo de token
auth_token='COPIE SU TOKEN AQUÍ' #donde dice COPIE SU TOKEN AQUI reemplace por el código alfanumério (token) obtenido en la registración
head = {'Authorization': 'Bearer ' + auth_token}
```

```python
#Por ejemplo, se consulta la serie de cotización del dólar
url = 'https://api.estadisticasbcra.com/usd'
response = requests.get(url, headers=head)
```

```python
#se obtienen los datos en formato json
json_data = response.json()
```

```python
#se convierte a DataFrame
df=pd.DataFrame(json_data)
df
#df.sort_values(by="d",ascending=False) opcional
```

```python
#se agrega nombre a las columnas
df.columns=['Fecha', 'Cotización del Dolar']
df
```
