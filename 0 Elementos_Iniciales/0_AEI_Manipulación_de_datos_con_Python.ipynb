Siempre se debe verificar lo realizado

```python
control=[]
for i,row in DATA_1[['EDAD','EDAD_2']].iterrows():
  if row.EDAD <= 30  and row.EDAD_2 == 'MENOS_30':
    control.append('OK')
  elif row.EDAD > 30 and row.EDAD <= 50 and row.EDAD_2 == 'ENTRE_31_50':
      control.append('OK')
  elif row.EDAD > 50 and row.EDAD <=70 and row.EDAD_2 == 'ENTRE_51_70':
      control.append('OK')
  else :
      control.append('ERROR')

import numpy as np
np.unique(control) #observamos que el único valor es OK
```

Las reglas definidas son a modo de ejemplo

```python
for i, row in DATA_1.iterrows():
  if row.OCUPACION == 'Student' and  row.MOROSIDAD_ULTIMOS_2AÑOS >= 2 :
    DATA_1.at[i,'RIESGO'] = 'ALTO'
  elif row.OCUPACION == 'Self-Employed' and row.MOROSIDAD_ULTIMOS_2AÑOS >= 3 :
    DATA_1.at[i,'RIESGO'] = 'ALTO'
  elif row.OCUPACION == 'Employed' and row.MOROSIDAD_ULTIMOS_2AÑOS >= 4 :
    DATA_1.at[i,'RIESGO'] = 'ALTO'
  elif row.MOROSIDAD_ULTIMOS_2AÑOS > 0 and row.MOROSIDAD_ULTIMOS_2AÑOS < 2:
    DATA_1.at[i,'RIESGO'] = 'MEDIO'
  else :
    DATA_1.at[i,'RIESGO'] = 'NORMAL'
```

```python
DATA_1['RIESGO'].unique()
```

Siempre se debe verificar lo realizado

```python
control_2=[]
for i,row in DATA_1[['OCUPACION','MOROSIDAD_ULTIMOS_2AÑOS','RIESGO']].iterrows():
  if row.OCUPACION == 'Student' and  row.MOROSIDAD_ULTIMOS_2AÑOS >= 2 and row.RIESGO == 'ALTO':
    control_2.append('OK')
  elif row.OCUPACION == 'Self-Employed' and row.MOROSIDAD_ULTIMOS_2AÑOS >= 3 and row.RIESGO == 'ALTO':
    control_2.append('OK')
  elif row.OCUPACION == 'Employed' and row.MOROSIDAD_ULTIMOS_2AÑOS >= 4 and row.RIESGO == 'ALTO':
    control_2.append('OK')
  elif row.MOROSIDAD_ULTIMOS_2AÑOS > 0 and row.MOROSIDAD_ULTIMOS_2AÑOS < 2 and row.RIESGO == 'MEDIO':
    control_2.append('OK')
  elif row.RIESGO == 'NORMAL':
    control_2.append('OK')
  else :
      control_2.append('ERROR')

np.unique(control_2) #observamos que el único valor es OK
```
