         I_t^{(Ag.Sim)} = {\sum_{k=1}^n p_{k;t}\over \sum_{k=1}^n p_{k;0}}*100

####**Índice de Precio Agregado Ponderado**



\begin{align}
         I_t^{(Ag.Pond)} = {\sum_{k=1}^n q_k*p_{k;t}\over \sum_{k=1}^n q_k*p_{k;0}}*100
\end{align}

```python
data= data
cols= [1,2]
cants= [20,30]
fila1= 21
filabase1=0

valornum=[]
for i,k in zip(cols,cants):
  for j in data.iloc[fila1:fila1+1,i]:
    valornum.append(j*k)
numerador=sum(valornum)
valordenom=[]
for i,k in zip(cols, cants):
  for j in data.iloc[filabase1:filabase1+1,i]:
    valordenom.append(j*k)
denominador=sum(valordenom)
IPAP=(numerador/denominador)*100
IPAP
```

####**Índice de Laspeyres**



\begin{align}
         IL_t = {\sum_{k=1}^n q_{i,0}*p_{i;t}\over \sum_{k=1}^n q_{i,0}*p_{i;0}}*100
\end{align}

```python
pd.set_option("display.precision", 14) #visualizar cantidad de decimales completa
dataall=pd.DataFrame(yf.download(['GOOG', 'AMZN'], start='2024-08-01', end='2024-09-01')[['Close', 'Volume']]).reset_index()
dataall
```

```python
data= dataall
colsp= [1,2]
colsq= [3,4]
fila1= 21
filabase1=0

valornum=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[fila1:fila1+1,i]:
    valornum.append(j*data.iloc[filabase1:filabase1+1,k])
numerador=sum(valornum)
valordenom=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[filabase1:filabase1+1,i]:
    valordenom.append(j*data.iloc[filabase1:filabase1+1,k])
denominador=sum(valordenom)
IL=list((numerador/denominador)*100)
IL[0]
```

####**Índice de Paschee**



\begin{align}
         IP_t = {\sum_{k=1}^n q_{i,t}*p_{i;t}\over \sum_{k=1}^n q_{i,t}*p_{i;0}}*100
\end{align}

```python
data= dataall
colsp= [1,2]
colsq= [3,4]
fila1= 21
filabase1=0

valornum=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[fila1:fila1+1,i]:
    valornum.append(j*data.iloc[fila1:fila1+1,k])
numerador=sum(valornum)
valordenom=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[filabase1:filabase1+1,i]:
    valordenom.append(j*data.iloc[fila1:fila1+1,k])
denominador=sum(valordenom)
IP=list((numerador/denominador)*100)
IP[0]
```

####**Índice de Fisher**



\begin{align}
         IF_t = \sqrt {IL_t * IP_t}
\end{align}

```python
data= dataall
colsp= [1,2]
colsq= [3,4]
fila1= 21
filabase1=0

valornum=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[fila1:fila1+1,i]:
    valornum.append(j*data.iloc[filabase1:filabase1+1,k])
numerador=sum(valornum)
valordenom=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[filabase1:filabase1+1,i]:
    valordenom.append(j*data.iloc[filabase1:filabase1+1,k])
denominador=sum(valordenom)
IL=list((numerador/denominador)*100)

valornum1=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[fila1:fila1+1,i]:
    valornum1.append(j*data.iloc[fila1:fila1+1,k])
numerador1=sum(valornum1)
valordenom1=[]
for i,k in zip(colsp, colsq):
  for j in data.iloc[filabase1:filabase1+1,i]:
    valordenom1.append(j*data.iloc[fila1:fila1+1,k])
denominador1=sum(valordenom1)
IP=list((numerador1/denominador1)*100)

IF=(IL[0]*IP[0])**0.5
IF
```

Se verifica este resultado del siguiente modo:

Dados $IL_{21}=96.74561220724019$ y $IP_{21}=96.62366959238139$

$IF_{21} = \sqrt {IL_{21} * IP_{21}} = \sqrt {96.74561220724019 * 96.62366959238139}= 96.68462167493357$

####**Cambio de base**



\begin{align}
         I_{baseY, t} = {I_{baseX, t}\over I_{baseX, Y}} *100
\end{align}

```python
pd.set_option("display.precision", 14) #visualizar cantidad de decimales completa
data=pd.DataFrame(yf.download(['GOOG', 'AMZN'], start='2024-08-01', end='2024-09-01')['Close']).reset_index()
data
```

Se crea un nuevo DataFrame directamente con los indice de precio (utilizando la función IPrecio creada previamente) de cada activo. En el casdo de AMZN se toma como período base 1/8/2024 y en el caso de GOOG 23/8/2024

```python
pd.set_option("display.precision", 14)
INDICES=pd.DataFrame({'IP_AMZN': valoresA, 'IP_GOOG': valoresG})
INDICES
```

Como los índices tienen período base diferentes, para que sean comparables, se realiza el cambio de base para AMZN, tomando como nuevo período base 23/8/2024

Se crea una función para recalcular un índice cambiando la base

```python
data= INDICES
col1= 0
filabasenew=16

valores=[]
for i in data.iloc[:,col1]:
  valores.append((i/data.iloc[filabasenew, col1])*100)

INDICES['IP_AMZN_NB']=valores
INDICES
```

Se verifica este resultado para algunos casos del siguiente modo:

$I_{16, 0} = {I_{16, 0}\over I_{16, 16}} *100 = {100.0\over 96.18079330778568} *100 = 103.97086212420039$
